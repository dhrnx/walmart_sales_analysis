```python
# ==========================================================
# ðŸ§¹ DATA CLEANING - Walmart Sales Dataset
# ==========================================================

import pandas as pd

# Load the dataset
df = pd.read_csv('../data/raw/Walmart_Store_sales.csv')

# Convert Date column (fix format issues)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Drop rows where Date couldn't be parsed (if any)
df = df.dropna(subset=['Date'])

# ==========================================================
# ðŸ§¾ Handle Missing Values
# ==========================================================
print("\nMissing values before cleaning:")
print(df.isnull().sum())

# Option 1: Drop rows with missing Weekly_Sales (if any)
df = df.dropna(subset=['Weekly_Sales'])

# Option 2: Fill numeric columns with mean (for simplicity)
num_cols = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

# ==========================================================
# âš™ï¸ Data Type Checks
# ==========================================================
print("\nData types before adjustments:")
print(df.dtypes)

# Ensure correct types
df['Store'] = df['Store'].astype(int)
df['Holiday_Flag'] = df['Holiday_Flag'].astype(int)
df['Weekly_Sales'] = df['Weekly_Sales'].astype(float)

# ==========================================================
# ðŸš« Handle Outliers (optional)
# ==========================================================
# Cap extreme Weekly_Sales values at 99th percentile
upper_limit = df['Weekly_Sales'].quantile(0.99)
df.loc[df['Weekly_Sales'] > upper_limit, 'Weekly_Sales'] = upper_limit

# ==========================================================
# ðŸ’¾ Save Cleaned Data
# ==========================================================
output_path = '../data/processed/walmart_sales_clean.csv'
df.to_csv(output_path, index=False, encoding='utf-8')
print(f"\nâœ… Cleaned dataset saved successfully at: {output_path}")

# Verify
print("\nCleaned dataset shape:", df.shape)
df.head()
```
